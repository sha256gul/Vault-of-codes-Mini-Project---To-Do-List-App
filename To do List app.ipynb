```python
import tkinter as tk
from tkinter import messagebox

tasks = []

def update_task_list():
    listbox.delete(0, tk.END)
    for i, task in enumerate(tasks):
        status = "✓" if task["completed"] else "✗"
        listbox.insert(tk.END, f"{i+1}. [{status}] {task['task']}")

def add_task():
    task_text = entry.get()
    if task_text:
        tasks.append({"task": task_text, "completed": False})
        entry.delete(0, tk.END)
        update_task_list()
    else:
        messagebox.showwarning("Input Error", "Please enter a task.")

def delete_task():
    try:
        selected_index = listbox.curselection()[0]
        tasks.pop(selected_index)
        update_task_list()
    except IndexError:
        messagebox.showwarning("Selection Error", "Please select a task to delete.")

def mark_complete():
    try:
        selected_index = listbox.curselection()[0]
        tasks[selected_index]['completed'] = True
        update_task_list()
    except IndexError:
        messagebox.showwarning("Selection Error", "Please select a task to mark complete.")

root = tk.Tk()
root.title("To-Do List App")

frame = tk.Frame(root)
frame.pack(pady=10)

entry = tk.Entry(frame, width=30)
entry.pack(side=tk.LEFT, padx=5)

add_button = tk.Button(frame, text="Add Task", command=add_task)
add_button.pack(side=tk.LEFT)

listbox = tk.Listbox(root, width=50)
listbox.pack(pady=10)

btn_frame = tk.Frame(root)
btn_frame.pack()

delete_button = tk.Button(btn_frame, text="Delete Task", command=delete_task)
delete_button.pack(side=tk.LEFT, padx=5)

complete_button = tk.Button(btn_frame, text="Mark Complete", command=mark_complete)
complete_button.pack(side=tk.LEFT)

root.mainloop()
```
